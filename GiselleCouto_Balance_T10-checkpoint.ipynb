## Sobre o Dataset:
→ Número de instâncias: 40690 para campaings.csv.
→ Número de atributos: 16 + atributo de saída.

→ Variáveis de entrada:

##### Atributos específicos do cliente:

0. Índice
1. Idade(age) : numérico
2. Emprego (job): tipo de trabalho (categórico: "admin.", "Desconhecido", "desempregado", "gerência", "empregada doméstica", "empreendedor", "estudante", "trabalhadores autônomos", "aposentados", "técnicos", "serviços")
3. Estado Civil (marital): estado civil (categórica: "casado", "divorciado", "solteiro"; nota: "divorciado" significa divorciado ou viúvo)
4. educação (education): (categórica: "desconhecido", "secundário", "primário", "terciário")
5. padrão (default): tem crédito em padrão? (binário: "sim", "não")
6. saldo (balance): saldo médio anual, em euros (numérico)
7. habitação (housing): tem empréstimo à habitação? (binário: "sim", "não")
8. empréstimo(loan): tem empréstimo pessoal? (binário: "sim", "não")

##### relacionados ao último contato da campanha atual:
9. contato(contact): tipo de comunicação de contato (categórico: "desconhecido", "telefone", "celular")
10. dia (day): dia do mês do último contato (numérico)
11. mês(month): mês do último contato do ano (categórico: "jan", "feb", "mar", ..., "nov", "dec")
12. duração (duration): duração do último contato, em segundos (numérico)

##### outros atributos:
13. campanha (campaing): número de contatos realizados durante esta campanha e para este cliente (numérico, inclui o último contato)
14. pdays: número de dias que passaram após o último contato do cliente de uma campanha anterior (numérico, .1 significa que o cliente não foi contatado anteriormente)
15. anterior (previous): número de contatos realizados antes desta campanha e para este cliente (numérico)
16. poutcome: resultado da campanha de marketing anterior (categórica: "desconhecido", "outro", "falha", "sucesso")

→ Variável de saída (destino desejado):
17. y - o cliente assinou um depósito a prazo? (binário: "sim", "não")

### Processamento de dados

transformando variáveis categórias em numéricas

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder
```

```python
le = LabelEncoder()
df.job=le.fit_transform(df.job)
df.marital = le.fit_transform(df.marital)
df.education =le.fit_transform(df.education)
df.default = le.fit_transform(df.default)
df.housing = le.fit_transform(df.housing)
df.loan = le.fit_transform(df.loan)
df.contact = le.fit_transform(df.contact)
df.month  = le.fit_transform(df.month)
df.poutcome = le.fit_transform(df.poutcome)
df.y=le.fit_transform(df.y)
```
